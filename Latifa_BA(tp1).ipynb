# Compte Rendu (TP1)

```python
def horner(a,z_0):
    n=len(a)
    result=a(1)
    for j in range(2, n+1):
        result=(result*z_0)+a(j)
    x=print(result)
```

```python
p= lambda x:(x-1)**6
f= lambda x:x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

```python
from numpy import linspace
T=[1, -6, 15, -20, 15, -6, 1]
for i in range((linspace(0.998,1.002,41))) :
    print (p(i))
print('----------------')
for i in range((linspace(0.998,1.002,41))) :
    print (f(i))
print('----------------')
for i in range((linspace(0.998,1.002,41))) :
    print (horner(T,i))
```

```python
import matplotlib.pyplot as plt
from numpy import linspace
T=linspace(0.998,1.002,41)
plt.plot(T,f(T))
plt.title("solution dans [0.998:1.002] ");
plt.xlabel("x")
plt.ylabel ("f(x)")
plt.grid(True)
```
