<h5>Affichage de nombre d'itérations effectuées :</h5>

```python
def dichotomiew (f ,a ,b , epsilon ):
    xa , xb = min (a , b ) , max (a , b )
    valxa , valxb = f ( xa ) , f ( xb )
    n = 0
    while xb - xa > epsilon :
        n += 1
        m = ( xa + xb )/2
        valm = f ( m )
        if valxa * valm < 0 :
            xb = m
            valxb = valm
        else :
            xa = m
            valxa = valm
    print (" nombre d'itérations :", n )
    return ( xa + xb )/2
```

<h3>Méthode numérique point fixe avec un codepython</h3>

```python
from sympy import *
import numpy as np
f=lambda x:  x**3+4*x**2-10 # la fonction f
g1=lambda x: 1/2*sqrt(10-x**3)
g2=lambda x: sqrt(10/(x+4))
g3=lambda x: x-x**3-4*x**2 + 10


def point_fixe(g,x0,eps,n):
    
    n=0
    x1=g(x0)
    while abs(x1-x0)>eps:
        n+=1
        x0=x1
        x1=g(x0)
    print("nombre d'iteration", n)
    return x1
  
```

```python
x1= point_fixe(g1,1.5,0.001,30)
print('la valeur approché x1=',x1)
x2= point_fixe(g2,1.5,0.001,30)
print('la valeur approché x2=',x2)
x3= point_fixe(g3,1.5,20,30)
print('la valeur approché x3=',x3)
```
