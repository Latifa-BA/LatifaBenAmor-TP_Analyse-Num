### Exercice 1 :

1) la fonction PR permet de résoudre Yi qui est égal à f(Xi)

```python
from numpy import prod
def PR(x,y,c):
    s=[]
    n=len(x)
    
    for j in range (n):
        
        L = [c-xx for xx in x]
        for k in range(n):
            if (k == j):
                L[k]= 1
            else:
                L[k] = L[k]/(x[j] - x[k])
                
        s.append(y[j] * prod(L))
        
    return sum(s)
```

```python
x=[1,2,3,4]
y=[-1,0,2,1]
for i in x:
    print(PR(x,y,i))
```

```python
def intN(x,y):
    m=len(x)
    
    np.copy(y)
    f=np.zeros(m)
    for k in range(1,m):
        f[k:m] = np.array(np.array(y[k:m+1])-np.array(y[k-1:m-1]))/np.array(np.array(x[k:m+1])-np.array(x[0:m-k]))
    return f
```

```python
x=[-1,0,2,5]
y=[6,1,3,66]
print(intN(x,y))
```

```python
def intNval(x,y,t):
    m=len(x)
    p=y[m-1]*np.ones(len(t))
    for k in range(m-2,-1,-1):
        p=p*np.array((np.array(t)-x[k]))+y[k]
    return p
```
